```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

## 2. Read in images or video

```python
IMAGE_PATH = './Downloads/img3.png'
```

```python
reader = easyocr.Reader(['en'], gpu = False)
result = reader.readtext(IMAGE_PATH)
result
```

## 3. Draw Results

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(IMAGE_PATH)
img = cv2.rectangle(img, top_left, bottom_right, (0,255,0),3)
img = cv2.putText(img,text,top_left,font,.9,(52,53,55),2,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

## 4. Handling multiple lines

```python
def readText(IMAGE_PATH):
    reader = easyocr.Reader(['en'], gpu = False)
    result = reader.readtext(IMAGE_PATH)
    return result
```

```python
def applyText(path):
    result  = readText(path)
    img = cv2.imread(path)
    for detection in result:
        top_left = tuple([val for val in detection[0][0]])
        bottom_right = tuple([val for val in detection[0][2]])
        text = detection[1]
        font = cv2.FONT_HERSHEY_SIMPLEX
        img = cv2.rectangle(img, top_left, bottom_right, (0,255,0),3)
        img = cv2.putText(img,text,top_left,font,.9,(220,20,60),2,cv2.LINE_AA)
    plt.figure(figsize=(10,10))
    plt.imshow(img)
    plt.show()
```

```python
applyText("./Downloads/img1.png")
```

```python
applyText("./Downloads/img2.jpg")
```
